```python
#Dropping many unrequired columns from the data frame
D = df[['Fl Date','Month','Origin',"Origin State Nm", 'Dest', "Dest State Nm", "Delay Arrival",'Delay Departures', 'Arr Delay Group','Route','Dep Delay Group','Node','Distance Group','Nas Delay','Security Delay','Weather Delay','Late Aircraft Delay','Carrier Delay','Cancelled']]
D[:5]
```

```python
#Replacing the null values with zero, as we cannot drop the rows, important for analysis 
D = D.fillna(0)
```

```python
#Creating a new coloumn in Data Frame so that we can group with both Month and Year
D['year_month'] = pd.to_datetime(D['Fl Date']).dt.to_period('M')
D[:5]
```

```python
#Tree Map to understand which state has the most departure delays
A = D.groupby("Origin State Nm")["Delay Departures"].mean().to_frame(name = "Mean Departure Delay in mins").reset_index()
A = A.sort_values("Mean Departure Delay in mins", ascending= False)
A
```

```python
norm = matplotlib.colors.Normalize(vmin=min(A["Mean Departure Delay in mins"]), vmax=max(A["Mean Departure Delay in mins"]))
colors = [matplotlib.cm.Blues(norm(value)) for value in A["Mean Departure Delay in mins"]]
```

```python
fig = plt.figure(figsize=(30,20))
squarify.plot(sizes= A["Mean Departure Delay in mins"], label=A["Origin State Nm"], color = colors, alpha=1, pad = True, text_kwargs={'fontsize':16})
plt.axis('off')
plt.show()
```

```python
#Tree Map to understand which state as the most arrival delays
B = D.groupby("Dest State Nm")["Delay Arrival"].mean().to_frame(name = "Mean Arrival Delay in mins").reset_index()
B = B.sort_values("Mean Arrival Delay in mins", ascending= False)
B
```

```python
norm = matplotlib.colors.Normalize(vmin=min(B["Mean Arrival Delay in mins"]), vmax=max(B["Mean Arrival Delay in mins"]))
colors = [matplotlib.cm.Reds(norm(value)) for value in B["Mean Arrival Delay in mins"]]
```

```python
fig = plt.figure(figsize=(30,20))
squarify.plot(sizes= B["Mean Arrival Delay in mins"], label=B["Dest State Nm"], color = colors, alpha=1, pad = True, text_kwargs={'fontsize':16})
plt.axis('off')
plt.show()
```
